```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
```

```python
#Uploading the CSV file
df = pd.read_csv("C:\\Users\\lavle\\Desktop\\Open food facts.csv", encoding = 'utf8')
#Remove products which do not have any product code(key field)
df  = df[df.code.isnull() == False]
df.isnull().sum()
df.head()
```

```python
#Convert code field into interger type as it has been converted to float
df.code = df.code.astype(int)
#Conversion successful
df.head()
df.sort_values(by='code')
df.head()
```

```python
list(df.columns)
```

```python
df = df[['code',
'created_datetime',
'last_modified_datetime',
'product_name',
'quantity',
'packaging_tags',
'brands',
'brands_tags',
'categories_en',
'origins_tags',
'manufacturing_places_tags',
'first_packaging_code_geo',
'purchase_places',
'countries_en',
'ingredients_text',
'serving_size',
'main_category_en',
'energy_100g',
'fat_100g',
'carbohydrates_100g',
'sugars_100g',
'fiber_100g',
'proteins_100g',
'calcium_100g',
'iron_100g',
'carbon-footprint_100g']]
df.shape
```

```python
import datetime
```

```python
df.created_datetime = pd.to_datetime(df['created_datetime'])
df.head()
```

```python
df['year'] = df['created_datetime'].dt.year
df['month'] = df['created_datetime'].dt.month
df = df[df.year.isnull() == False]
df = df[df.month.isnull() == False]
df.year = df.year.astype(int)
df.month = df.month.astype(int)
df.head()
```

```python
del df['last_modified_datetime']
```

```python
df_countries = df[['code',
'product_name',
'quantity',
'packaging_tags',
'brands',
'brands_tags',
'categories_en',
'origins_tags',
'manufacturing_places_tags',
'first_packaging_code_geo',
'purchase_places',
'countries_en',
'energy_100g',
'fat_100g',
'carbohydrates_100g',
'sugars_100g',
'fiber_100g',
'proteins_100g',
'calcium_100g',
'iron_100g',
'carbon-footprint_100g',
'year',
'month']]
df_countries.shape
```

```python
df_countries = df_countries[df_countries['countries_en'].isnull() == False]
```

```python
country_lc = df_countries.countries_en.str.split(',', expand=True).stack()
country_llc = pd.Series.to_frame(country_lc)
country_llc.columns = ['country']
country_llc.head()
```

```python
df_country, df2_country = df_countries.align(country_llc, level=0, join='outer')
df_country[139:147]df_country[139:147]
```

```python
df_country['country'] = df2_country['country']
df_country.head()
```

```python
del df_country['countries_en']
df_country.head()
```

```python
df_country.reset_index(drop=True,inplace=True)
df_country[139:147]
```

```python
country = ['Switzerland', 'Belgium','Spain', 'Germany', 'United Kingdom', 'Guandeloupe']#, 'France'
df_cntry_1 = df_country[df_country.country.isin(country)]
df_cntry_1 = df_cntry_1[df_cntry_1.country.isnull() == False]
df_cntry_1.sort_values(by='country', inplace=True)
df_cntry_1['country'].value_counts().plot(kind='bar')
plt.show()
```

```python
years = ['2015', '2016', '2017']
df_cntry_2 = df_cntry_1[df_cntry_1.year.isin(years)]
df_cntry_2.sort_values(by='year', inplace=True)
df_cntry_2['year'].value_counts().plot(kind='bar')
plt.show()
```

```python
search = ['plastique', 'plastic']
df_cntry_3 = df_cntry_2[df_cntry_2.packaging_tags.notnull()]
df_cntry_3 = df_cntry_3[df_cntry_3['packaging_tags'].str.contains('|'.join(search))]
df_cntry_3.head()
```

```python
df_cntry_3.shape
```

```python
from pylab import rcParams
```

```python
color = ['pink', 'grey', 'green']
c=0
rcParams['figure.figsize'] = 13, 7
fig, ax = plt.subplots()
#for i in range(2015, 2018):
plastic_c = np.array(df_cntry_3.country[df_cntry_3.year == 2015].value_counts(sort=False))
N = plastic_c.shape[0]
ind = np.arange(N)*2.5 
width = 0.3
p1 = plt.bar(ind + width*c*1, plastic_c, width, color= color[c])

c+= 1
    
plastic_c = np.array(df_cntry_3.country[df_cntry_3.year == 2016].value_counts(sort=False))
N = plastic_c.shape[0]
ind = np.arange(N)*2.5 
width = 0.3
p2 = plt.bar(ind + width*c*1, plastic_c, width, color= color[c])

c+=1

plastic_c = np.array(df_cntry_3.country[df_cntry_3.year == 2017].value_counts(sort=False))
N = plastic_c.shape[0]
ind = np.arange(N)*2.5 
width = 0.3
p3 = plt.bar(ind + width*c*1, plastic_c, width, color= color[c])


plt.ylabel('Plastic Consumption', fontsize='large')
plt.xlabel('Countries', fontsize='large')

plt.title('Plastic consumption over the years 2015-2017', fontsize='large')
for i in range(2015, 2018):
    plt.xticks(ind+width, df_cntry_3.country[df_cntry_3.year == i].value_counts(sort= False).index, rotation = 45, fontsize='large')
plt.legend((p1, p2, p3), ('2015', '2016', '2017'), fontsize='large')
#c+= 1
plt.show()
```
